#### Standardization 
- There is a formula for standardization,  
 
Standardized value have mean as 0 and std deviation of 1




```python
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill=list(df['total_bill'])
```

```python
sns.kdeplot(total_bill)
```

```python
import pandas as pd

total_bill1 = pd.DataFrame(total_bill)
```

```python
total_bill1.skew()
```

```python
total_bill
```

- There are two ways, we can standardize the data
1. Manual method 
2. Using library's inbuilt function 

```python
# manual 
import numpy as np
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std ## z_score
    normalized_data.append(z_score)

print(normalized_data)
```

```python
sns.kdeplot(normalized_data)
```

#### Using Sklearn Library 


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill', 'tip']]) # first we have to fit the data and then we can transform it 
```

```python
std_data = pd.DataFrame(scaler.transform(df[['total_bill', 'tip']]), columns=['std_bill', 'std_tip'])
```

## Normalization 
- Normalization means scaling the data in the range of 0 to 1
- The formula to achieve this is the following: Min-max normalization preserves the relationships among the original data values.

- Standardization genrally used for ML problems, and min-max scaler genrally used for DL problems 





```python
df.head()
```

```python
# lets apply the min-max scaler on total bill column and tip column 

tip2 = df.tip
total_bill2 = df.total_bill
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
# First we need to fit the data or we can fit and transform the data simultaneously 
min_max.fit(df[['total_bill', 'tip']])
```

```python
min_max.transform(df[['total_bill', 'tip']])
```

```python
# to save this as data frame we can use pandas function to save it as a data frame 

transformed_bill_tip = pd.DataFrame(min_max.fit_transform(df[['total_bill', 'tip']]), columns=['total_bill', 'tip'])
```
